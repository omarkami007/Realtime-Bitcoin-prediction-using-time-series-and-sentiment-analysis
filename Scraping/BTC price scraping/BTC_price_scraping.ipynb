```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta
```

```python
# Parameters:
ndays = 30 #30 days by default (start date: today -30 days, end date: today)
time_step = 60   # 60 minutes is default value
```

# Scrape BTC price following parameters

```python
END_DATE = datetime.now().strftime('%Y-%m-%d')  # Aujourd'hui
START_DATE = (datetime.now() - timedelta(days=ndays)).strftime('%Y-%m-%d')

# Paramètres de configuration
STOCK_SYMBOLS = ['BTC-USD']  # Symboles des actions à analyser


# Collecter les données boursières
def collect_stock_data(symbol, start_date, end_date):
    stock_data = yf.download(symbol, start=start_date, end=end_date, interval=f'{int(time_step)}m')
    stock_data['Symbol'] = symbol
    return stock_data

stock_df = collect_stock_data('BTC-USD',start_date=START_DATE,end_date=END_DATE)

stock_df
```

# Save BTC data to csv

```python
stock_df.to_csv('btc_data.csv',encoding='utf-8-sig',index= False)
```
